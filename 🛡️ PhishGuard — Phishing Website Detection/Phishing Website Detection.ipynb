```python
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer(r"[A-Za-z]+")
```

```python
df.URL[0]
```

```python
tokenizer.tokenize(df.URL[0])
```

```python
df["Text Tokenized"] = df["URL"].map(lambda t: tokenizer.tokenize(t))
```

```python
df.head()
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
stemmer = SnowballStemmer('english')
df["Text Stemmed"] = df['Text Tokenized'].map(lambda t: [stemmer.stem(word) for word in t])
```

```python
df.head()
```

```python
df["Text"] = df["Text Stemmed"].map(lambda l: " ".join(l))
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
df.head()
```

```python
features = cv.fit_transform(df["Text"])
```

```python
features[:5].toarray()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test , y_train , y_test = train_test_split(features , df["Label"] , test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train , y_train)
```

```python
model.score(x_test , y_test)
```

Our model gives the accuracy of about 96%

```python
model.score(x_train , y_train)
```

Hence the model is not overfitted
